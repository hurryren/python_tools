```python
import matplotlib.pyplot as plt
```

```python
def cal_time(pkt_size=64):
    return pkt_size * (8 + 5) * 2 / 1000.0
    
```

```python
x = [64,128,256,512,1024,1280,1514]
th_time = [cal_time(t) for t in x ]
print(th_time)
```

```python
fact_time = [5.6,7.77,12.11,20.80,38.8,46.87,54.95]
loop_time = [2.40,2.91,3.93,5.97,10.08,12.11,14.01]
th_loop = []
for index,y in enumerate(th_time):
    th_loop.append(y+loop_time[index])
th_loop
```

```python
# 这两行代码使得 pyplot 画出的图形中可以显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


plt.plot(x,fact_time,'.-',label="实际延时")
plt.plot(x,th_time,'.-',label="理论延时")
plt.plot(x,loop_time,'.-',label="环回延时")
plt.plot(x,th_loop,'.-',label="环回 + 理论")
plt.xlabel("包大小(Byte)")
plt.ylabel("延时(us)")
plt.legend()
plt.grid()
plt.show()
```
